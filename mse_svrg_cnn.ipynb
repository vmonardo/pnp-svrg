```python
params = create_problem(img_path='./data/mse_images/tilt_008.tiff')

psnr_display(img_path='./data/mse_images/tilt_008.tiff', output=params['original'], title="Original")
psnr_display(img_path='./data/mse_images/tilt_008.tiff', output=params['noisy'], title="Noisy")

svrg_out, svrg_time, svrg_psnr, svrg_zs = pnp_svrg(params=params,
                                                   denoiser='cnn',
                                                   eta=20e3, tt=20, T2=10, 
                                                   mini_batch_size=int(20e3), verbose=False)

psnr_display(img_path='./data/mse_images/tilt_008.tiff', output=svrg_out, title="SVRG")
```
